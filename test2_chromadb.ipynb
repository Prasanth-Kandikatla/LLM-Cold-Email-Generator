```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="my_collection")
```

```python
collection.add(
    ids=["id1", "id2"],
    documents=[
        "This is a document about San Francisco",
        "This is a document about Hyderabad"
    ]
)
```

```python
all_docs = collection.get()
all_docs
```

```python
results = collection.query(
    query_texts=["This is a query document about charminar"], # Chroma will embed this for you
    n_results=2 # how many results to return
)
print(results)
```
